### Task 5: Web Scraping for Product Images

- *Objective*: Scrape product images from e-commerce websites for training data ``CNN_Model_Train_Data.csv``.
- *Key Actions*: Automate scraping, download images, and store them systematically and make sure you have enough data to train the CNN model.

```python
# Import the scraper
from src.utils.scraping_utils import ProductScraper

# Set up directories
DATA_DIR = os.path.join(project_root, 'src', 'data')
scraper = ProductScraper(DATA_DIR)

# Print directory info
print("Output directory:", DATA_DIR)
print("Directory exists:", os.path.exists(DATA_DIR))
print("Directory contents before scraping:", os.listdir(DATA_DIR))

# Test with a sample e-commerce site
test_url = "https://www.amazon.com/s?k=game+console&rh=n%3A7926841011&ref=nb_sb_noss"  # Replace with actual URL
try:
    print("\nStarting scraping...")
    scraper.scrape_product_images(test_url, max_images=10)
    
    # Verify results
    print("\nAfter scraping:")
    print("Directory contents:", os.listdir(DATA_DIR))
    
    # Check if images were downloaded
    image_files = [f for f in os.listdir(DATA_DIR) if f.endswith('.jpg')]
    print(f"\nDownloaded {len(image_files)} images:")
    for img in image_files:
        img_path = os.path.join(DATA_DIR, img)
        size = os.path.getsize(img_path)
        print(f"- {img} ({size} bytes)")
        
except Exception as e:
    print(f"Error during scraping: {str(e)}")
```
